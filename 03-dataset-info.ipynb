```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os.path import join
from matplotlib import pyplot as plt

dic = {
    0: 'empty',
    1: 'half',
    2: 'full'
}
```

# Import csv in DataFrame

```python
training_dataset_df = pd.read_csv(join('dataset', 'training.csv'))
validation_dataset_df = pd.read_csv(join('dataset', 'validation.csv'))
test_dataset_df = pd.read_csv(join('dataset', 'test.csv'))
dataset_df = pd.read_csv(join('dataset', 'all_labels.csv'))
```

```python
print('Occurency of all dataset')
print(dataset_df['label'].value_counts())
print('Total element %d' % (dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of training dataset')
print(training_dataset_df['label'].value_counts())
print('Total element %d' % (training_dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of validation dataset')
print(validation_dataset_df['label'].value_counts())
print('Total element %d' % (validation_dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of test dataset')
print(test_dataset_df['label'].value_counts())
print('Total element %d' % (test_dataset_df['label'].value_counts().sum()) )
print('---------------')
```

# Show some element of training dataset

## Training dataset

```python
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.title("Class: %s" % dic[training_dataset_df['label'][i]])
    plt.imshow(plt.imread(training_dataset_df['image'][i]),cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.title("Class: %s" % dic[validation_dataset_df['label'][i]])
    plt.imshow(plt.imread(validation_dataset_df['image'][i]))
plt.show()
```

```python
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.title("Class: %s" % dic[test_dataset_df['label'][i]])
    plt.imshow(plt.imread(test_dataset_df['image'][i]))
plt.show()
```

# Means and standard deviations using custom classes
Calculate mean and standard deviation for each channel of image

```python
from libs.TrashbinDataset import TrashbinDataset
```

```python
dataset_trbd = TrashbinDataset('dataset/all_labels.csv')

print('Len of dataset %d' % (dataset_trbd.__len__()))

means = np.zeros(3)
stdevs = np.zeros(3)

for data in dataset_trbd:
    img = data[0]
    for i in range(3):
        img = np.asarray(img)
        means[i] += img[i, :, :].mean()
        stdevs[i] += img[i, :, :].std()

means = np.asarray(means) / dataset_trbd.__len__()
stdevs = np.asarray(stdevs) / dataset_trbd.__len__()

print("{} : normMean = {}".format(type, means))
print("{} : normstdevs = {}".format(type, stdevs))
```
