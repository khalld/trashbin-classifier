```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os.path import join
```

```python
training_dataset_df = pd.read_csv(join('dataset', 'training.csv'))
validation_dataset_df = pd.read_csv(join('dataset', 'validation.csv'))
test_dataset_df = pd.read_csv(join('dataset', 'test.csv'))
dataset_df = pd.read_csv(join('dataset', 'all_labels.csv'))
```

```python
print('Occurency of all dataset')
print(dataset_df['label'].value_counts())
print('Total element %d' % (dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of training dataset')
print(training_dataset_df['label'].value_counts())
print('Total element %d' % (training_dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of validation dataset')
print(validation_dataset_df['label'].value_counts())
print('Total element %d' % (validation_dataset_df['label'].value_counts().sum()) )
print('---------------')

print('Occurency of test dataset')
print(test_dataset_df['label'].value_counts())
print('Total element %d' % (test_dataset_df['label'].value_counts().sum()) )
print('---------------')
```

Show some element of training dataset

```python
from matplotlib import pyplot as plt

dic = {
    0: 'empty',
    1: 'half',
    2: 'full'
}

plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.title("Class: %s" % dic[training_dataset_df['label'][i]])
    plt.imshow(plt.imread(training_dataset_df['image'][i]),cmap='gray')
plt.show()
```
